```python
non_num_feats = ["merchant", "category", "city", "state", "job"]

#encoding non-numeric features by replacing their values with their frequecies, since they have too many values to do one hot encode
for i in non_num_feats:
    # Calculate the frequency of each value of a feature
    feat_frequencies = df[i].value_counts()

    # Create a mapping dictionary from feature names to their frequencies
    feat_to_frequency = feat_frequencies.to_dict()

    # Replace the feature names with their frequencies in the DataFrame
    df[i + ' encoded'] = df[i].map(feat_to_frequency)
    
    
#categorize time into time of day    
df[['date', 'time']] = df['trans_date_trans_time'].str.split(' ', 1, expand=True)

# Convert the "time" column to a datetime object
df['time'] = pd.to_datetime(df['time'], format='%H:%M:%S')

# Define a function to categorize time of day
def categorize_time_of_day(time):
    if 23 <= time.hour or time.hour <5:
        return 'night/after midnight'
    elif 5 <= time.hour < 11:
        return 'morning'
    elif 11 <= time.hour < 17:
        return 'noon/afternoon'
    elif 17 <= time.hour < 23:
        return 'evening'

# Apply the categorize_time_of_day function to create a new "time_of_day" column
df['time_of_day'] = df['time'].apply(categorize_time_of_day)

#one-hot-encode time to time of day
encoded_df = pd.DataFrame()
one_hot_encoded = pd.get_dummies(df['time_of_day'], prefix='time_of_day')
encoded_df = pd.concat([encoded_df, one_hot_encoded], axis=1)
df = pd.concat([df, encoded_df], axis=1)


#create new feature "age" based on "date of birth"
from datetime import datetime

# Convert the "date_of_birth" column to datetime objects
df['dob'] = pd.to_datetime(df['dob'], format='%Y-%m-%d')

# Calculate the current date
current_date = datetime.now()

# Define a function to calculate age
def calculate_age(dob):
    age = current_date.year - dob.year - ((current_date.month, current_date.day) < (dob.month, dob.day))
    return age

# Apply the calculate_age function to create a new "age" column
df['age'] = df['dob'].apply(calculate_age)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate correlations
correlation_matrix = df.corr()

# Create a heatmap using seaborn
plt.figure(figsize=(30, 30))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
target_feature = "is_fraud"
correlations = correlation_matrix[target_feature]
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
sns.barplot(x=correlations.index, y=correlations.values)
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.xlabel("Features")
plt.ylabel("Correlation")
plt.title(f"Correlation between {target_feature} and other features")
plt.tight_layout()
plt.show()
```

```python
features = ['amt', 'lat', 'long', 'merch_long', 'merch_lat', 'age', 'state encoded']

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, accuracy_score

#import different ML algorithms
dtc = DecisionTreeClassifier()
rf = RandomForestClassifier()
lr = LogisticRegression()

#build features, target feature
X = df[features]
y = df.is_fraud

train_X, val_X, train_y, val_y = train_test_split(X, y, test_size = 0.2, random_state = 0)

#test basic model
compare = []
for model in (dtc, rf, lr):
    #train models
    model.fit(train_X,train_y)
    
    #make prediction
    prediction = model.predict(val_X)
    
    #calculate mean absolute error
    val_mae = mean_absolute_error(val_y, prediction)
    
    #calculate accuracy score
    ac_score = accuracy_score(val_y, prediction)
    
    compare.append({'model': model, 'mae': val_mae, 'ac_score': ac_score})

#compare results
table = pd.DataFrame(compare)
table
```

```python
#hyperparameter tuning
from sklearn.model_selection import GridSearchCV

# Define a grid of hyperparameters to search
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 5, 10, 15],
    'min_samples_leaf': [1, 5, 10],
    'min_samples_split': [2, 5, 10]
}

# Create a GridSearchCV object with cross-validation
grid_search = GridSearchCV(estimator= dtc, param_grid=param_grid, cv=5)

# Fit the grid search to the training data
grid_search.fit(train_X, train_y)

# Print the best hyperparameters
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)

# Get the best model from the grid search
best_model = grid_search.best_estimator_

# Evaluate the best model on the test data
accuracy = accuracy_score(val_y, best_model.predict(val_X))
print("Test Accuracy:", accuracy)
```

```python
results_df = pd.DataFrame(grid_search.cv_results_)
results_df
```

```python
# Sort the results by mean_test_score in descending order
sorted_results = results_df.sort_values(by='mean_test_score', ascending=False)

# Select the top 3 best models and their corresponding hyperparameters
top_3_models = sorted_results.head(3)
top_3_models
```
